# 5. Perform meaningful visualizations using the dataset. Bring atleast 3 visualizations.


# (1) Histogram

```python
plt.figure()
plt.hist(data['SL'])
plt.title("Distribution of Sepal Length")
plt.xticks()
plt.yticks()
plt.figure()
plt.hist(data['SW'],color='red')
plt.title("Distribution of Sepal Width")
plt.xticks()
plt.yticks()
plt.figure()
plt.hist(data['PL'],color='green')
plt.title("Distribution of Petal Length")
plt.xticks()
plt.yticks()
plt.figure()
plt.hist(data['PW'],color='yellow')
plt.title("Distribution of Petal Length")
plt.xticks()
plt.yticks()
```

# (2) Scatter Plot

```python
plt.figure(figsize=(8,8))
plt.scatter(data['SL'],data['SW'],s=30)
plt.title("Plot of Sepal Length vs Sepal Width",fontsize=16)
plt.xlabel('SL')
plt.ylabel('SW')
```

```python
plt.figure(figsize=(8,8))
plt.scatter(data['PL'],data['PW'],s=30,color='purple')
plt.title("Plot of Petal Length vs Petal Width",fontsize=16)
plt.xlabel('PL')
plt.ylabel('PW')
```

# (3)Pair Plots

```python
sns.pairplot(data)
```

# (4)Box Plot

```python
plt.figure(figsize=(15,10))
sns.boxplot(x='SL',y='SW',hue='Classification',data=data)
plt.title('Sepal Length vs Sepal Width',size=14)
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x='PL',y='PW',hue='Classification',data=data)
plt.title('Petal Length vs Petal Width',size=14)
```

# (5)Violin Plot

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Classification',y='PL',data=data)
plt.title('Petal Length of different Classification')
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Classification',y='SL',data=data)
plt.title('Sepal Length of different Classification')
```
